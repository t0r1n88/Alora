```python
"""
Скрипт для изменения структуры сложной экселевской таблицы.
"""
```

```python
import pandas as pd
```

```python
df =pd.read_excel('data/12132.xlsx',skiprows=18,header=None)
```

```python
df.head(5)
```

```python
out_df = pd.DataFrame(columns=['Наименование','Количество'])
```

```python
df.iloc[0,0]
```

```python
df.iloc[1,7]
```

```python
count = 0 # счетчик
for idx in range(len(df) + 1):
   
    temp_df = df.iloc[count:count+2,:]

    name = temp_df.iloc[0,0]
    value = temp_df.iloc[1,7]
    new_row = {'Наименование':name,'Количество':value}
    out_df = out_df.append(new_row, ignore_index=True)

    if name == 'Итого':
        break
    
    
    count += 2
```
