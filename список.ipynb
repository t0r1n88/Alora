```python
import pandas as pd
import string
import re
```

```python
df = pd.read_excel('data/prof.xlsx',dtype=str)
temp_df = pd.DataFrame(columns=df.columns)
```

```python
df.head()
```

```python
temp_df
```

```python
df['Профессия'] = df['Профессия'].fillna('Не заполнено')
```

```python
for row in df.itertuples():
    temp_row = list(row[1:])
    lst_value =re.sub(r'\d{2}.\d{2}.\d{2}[.]?','',row[4]).split(',')
    lst_value = list(map(str.strip,lst_value))
    lst_value = list(map(str.capitalize,lst_value))

    for prof in lst_value:
        new_row = pd.DataFrame(columns=temp_df.columns,data=[temp_row])
        new_row['Профессия'] = prof
        temp_df = pd.concat([temp_df,new_row],axis=0)


```

```python
temp_df
```

```python
temp_df.drop(columns=['Отметка времени'],inplace=True)
temp_df.rename(columns={'Выберите отрасль (направление) которую будете представлять':'Отрасль'})
```

```python
temp_df.to_excel('data/Профессии.xlsx',index=False,header=True)
```
