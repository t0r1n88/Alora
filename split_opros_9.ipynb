```python
import pandas as pd
import re
```

```python
def clean_end(value:str):
    
    value = value.strip()
    if value.endswith(','):
        return value[:-1]
    elif value.endswith('.'):
        return value[:-1]
    else:
        return value
```

```python
def split_column(value):
    if isinstance(value,str):
        value = re.findall(r'([ЁА-Я][а-я,-;.\(\)\s]+)',value)
        temp_lst = list(map(clean_end,value))
        value = '&'.join(temp_lst)
        return value
    else:
        return None
```
