```python
import pandas as pd
import re
```

```python
def add_zero_inn(cell):
    value = str(cell)
    if len(value) == 11:
        return f'0{value}'
    else:
        return value
    
```

```python
def process_snils(value):
    snils = str(value)
    result = re.findall(r'\d', snils) # ищем цифры
    if len(result) == 11:
        first_group = ''.join(result[:3])
        second_group = ''.join(result[3:6])
        third_group = ''.join(result[6:9])
        four_group = ''.join(result[9:11])

        out_snils = f'{first_group}-{second_group}-{third_group} {four_group}'
        return out_snils
    else:
        return ''

```

```python
def place_reborn(cell):
    value = str(cell)
    result = re.search(r'\b(пгт\s?[.]\s?[ЁёА-Яа-я-]+\W|г\s?[.]\s?[ЁёА-Яа-я-]+\W|у\s?[.]\s?[ЁёА-Яа-я-]+\W|с\s?[.]\s?[ЁёА-Яа-я-]+\W|'
                   r'п\s?[.]\s?[ЁёА-Яа-я-]+\W|п/ст\s?[.]\s?[ЁёА-Яа-я-]+\W)',value,re.IGNORECASE)
    if result:
       
        return result.group(0)
    else:
        return ''
```

```python
df = pd.read_excel('data/Список студентов.xlsx',dtype=str)
```

```python
df.fillna('',inplace=True)
```

```python
df['ИНН'] = df['ИНН'].apply(add_zero_inn)
df['СНИЛС'] = df['СНИЛС'].apply(process_snils)
df['Место_рождения'] = df['Адрес_регистрации'].apply(place_reborn)
```
